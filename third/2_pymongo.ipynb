```python
import json 
with open('/content/drive/MyDrive/Colab Notebooks/3rd project/engIndex.json') as json_file:
    word_index = json.load(json_file)

from keras.models import load_model
loaded_model = load_model('/content/drive/MyDrive/Colab Notebooks/3rd project/history/history10.h5')
```

```python
sentences = [sentence.split(' ') for sentence in want['text']]
tokenizer = Tokenizer(17666)
tokenizer.word_index = word_index
encoded = tokenizer.texts_to_sequences(sentences)
pad_new = pad_sequences(encoded, maxlen=30)
```

```python
predictions = loaded_model.predict(pad_new)
predict = pd.DataFrame(predictions)
```

```python
want['s'] = 0
for i in range(len(want)):
    want['s'][i] = predict.iloc[i].idxmax()
want['s'].value_counts()
```
